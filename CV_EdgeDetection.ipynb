```python
img_path = "/content/drive/MyDrive/brain MRI/yes/Y104.jpg"
img_path2="/content/drive/MyDrive/brain MRI/yes/Y114.JPG"
image = cv2.imread(img_path)
image2=cv2.imread(img_path2)
print("width: {} pixels".format(image.shape[1]))
print("height: {} pixels".format(image.shape[0]))
print("channels: {}".format(image.shape[2]))
dim=(500,590)
image=cv2.resize(image, dim)
image2=cv2.resize(image2,dim)
```

```python
plt.imshow(image2)
```

```python
img_blur = cv2.GaussianBlur(image,(5,5), 3)
plt.imshow(img_blur)
```

```python
img_blur2 = cv2.GaussianBlur(image2,(5,5), 3)
plt.imshow(img_blur2)
```

## **Sobel**

```python
sobelx=cv2.Sobel(src=img_blur,ddepth=cv2.CV_64F,dx=0,dy=1,ksize=5)#Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
```

```python
titles = ['Original Image', 'SobelX',
            'SobelY', 'SobelXY']
images = [image, sobelx, sobely, sobelxy]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
sobelx2=cv2.Sobel(src=img_blur2,ddepth=cv2.CV_64F,dx=0,dy=1,ksize=5)#Detection on the X axis
sobely2 = cv2.Sobel(src=img_blur2, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy2 = cv2.Sobel(src=img_blur2, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
```

```python
titles = ['Original Image', 'SobelX',
            'SobelY', 'SobelXY']
images = [image2, sobelx2, sobely2, sobelxy2]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

## Canny Edge Detector

```python
edges1 = cv2.Canny(image=img_blur, threshold1=100, threshold2=200)
# Display Canny Edge Detection Image
plt.imshow( edges1,cmap='gray')
```

```python
edges2 = cv2.Canny(image=img_blur2, threshold1=100, threshold2=200)
# Display Canny Edge Detection Image
plt.imshow( edges2,cmap='gray')
```

```python
titles = ['Original Image1', 'edges1',
            'Original Image2', 'edges2']
images = [image, edges1, image2, edges2]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

### Varying thresholds

```python
edges1b = cv2.Canny(image=img_blur, threshold1=50, threshold2=150)
# Display Canny Edge Detection Image
plt.imshow( edges1b,cmap='gray')
```

```python
edges1c = cv2.Canny(image=img_blur, threshold1=0, threshold2=100)
# Display Canny Edge Detection Image
plt.imshow( edges1c,cmap='gray')
```

```python
titles = ['Original Image', '100<threshold<200',
            '50<threshold<150', '0<threshold<100']
images = [image, edges1, edges1b, edges1c]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
